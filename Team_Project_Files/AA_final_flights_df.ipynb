```python
# Import Dependencies
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
```

```python
url = "https://flightprojectnorthwestern.s3.us-east-2.amazonaws.com/flights_clean.csv"

flight_df = pd.read_csv(url)

flight_df.head()
```

```python
# from 2019 - 2021 --> which airline seemed to lose the most $ based on ticket price
# or revenue drops from each year and quarter
# by carrier

revenue_grouped_df= flight_df.groupby(["Year"])["Ticket Cost"].sum()
revenue_grouped_df.plot(kind='bar', figsize=(12,5))
plt.xlabel(" Total Revenue Per Year")
plt.ylabel(" Total Revenue Gained (billions)")
plt.xticks(rotation=0, horizontalalignment='center')
plt.title(" Total Revenue")
plt.show()
```

```python
#visual top 5 airlines
revenue2019 = flight_df[flight_df["Year"] == 2019]
new_df = revenue2019[revenue2019.loc[:]!=0].dropna()
airlines_revenue2019= new_df.groupby(["Ticket Carrier"])["Ticket Cost"].sum().sort_values()
airlines_revenue2019.plot(kind='bar', figsize=(12,5))
plt.xlabel(" Total Revenue Per Airline (2019)")
plt.ylabel(" Total Revenue Gained")
plt.xticks(rotation=0, horizontalalignment='center')
plt.title(" Total Revenue (2019) ")
plt.show()
```

```python
revenue2020 = flight_df[flight_df["Year"] == 2020]
new_df_2020 = revenue2020[revenue2020.loc[:]!=0].dropna()
airlines_revenue2020= new_df_2020.groupby(["Ticket Carrier"])["Ticket Cost"].sum().sort_values()
airlines_revenue2020.plot(kind='bar', figsize=(12,5))
plt.xlabel(" Total Revenue Per Airline (2020)")
plt.ylabel(" Total Revenue Gained")
plt.xticks(rotation=0, horizontalalignment='center')
plt.title(" Total Revenue (2020) ")
plt.show()
```

```python
revenue2020_byq = flight_df[flight_df["Year"] == 2020]
rev_byq_2020 = revenue2020_byq[revenue2020_byq.loc[:]!=0].dropna()
revenuebyq2020= rev_byq_2020.groupby(["Quarter"]).sum()["Ticket Cost"]
revenuebyq2020.plot(kind='bar', figsize=(12,5))
plt.xlabel(" Total Revenue Per Quarter (2020)")
plt.ylabel(" Total Revenue Gained")
plt.xticks(rotation=0, horizontalalignment='center')
plt.title(" Total Revenue (2020) ")
plt.show()
```

```python
revenue2021_byq = flight_df[flight_df["Year"] == 2021]
rev_byq_2021 = revenue2021_byq[revenue2021_byq.loc[:]!=0].dropna()
revenuebyq2021= rev_byq_2021.groupby(["Quarter"]).sum()["Ticket Cost"]
revenuebyq2021.plot(kind='bar', figsize=(12,5))
plt.xlabel(" Total Revenue Per Quarter (2021)")
plt.ylabel(" Total Revenue Gained")
plt.xticks(rotation=0, horizontalalignment='center')
plt.title(" Total Revenue (2021) ")
plt.show()
```

```python
vaccinations_data_to_load = "Datasets/country_vaccinations.csv"
vaccinations_data=pd.read_csv(vaccinations_data_to_load)

vaccinations_data_usa = vaccinations_data[vaccinations_data["country"] == "United States"]
vaccinations_data_usa_grouped= vaccinations_data_usa.groupby(["date"]).sum()["total_vaccinations"]
vaccinations_data_usa_grouped.plot(kind='line', figsize=(12,5),color="red")
plt.xlabel(" USA Covid-19 Vaccinations ")
plt.ylabel(" Rate of Covid-19 Vaccinations Distribution ")
plt.xticks(rotation=0, horizontalalignment='center')
plt.title(" USA Population Vaccinated (2020,2021,2022) ")
plt.show()
```
